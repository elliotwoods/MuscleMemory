```python
from matplotlib import pyplot as plt, gcf
import numpy as np
import math
```

```python
cycle_resolution = 1024

i = np.arange(0, 1024)

sineTable = 255 * np.sin(math.pi * i / cycle_resolution)
plt.plot(sineTable)
```

```python
# Plot the A, B, C, D traces as an example

# This is not real data

phi = np.arange(0, math.pi * 2, 0.01)

A = np.cos(phi)
B = np.cos(phi - math.pi/2)
C = np.cos(phi - math.pi)
D = np.cos(phi - 3 * math.pi/2)

# These values now need to be rectified
A = np.clip(A, 0, 1)
B = np.clip(B, 0, 1)
C = np.clip(C, 0, 1)
D = np.clip(D, 0, 1)


plt.subplot(4, 1, 1)
plt.plot(phi / math.pi, A)

plt.subplot(4, 1, 2)
plt.plot(phi / math.pi, B)

plt.subplot(4, 1, 3)
plt.plot(phi / math.pi, C)

plt.subplot(4, 1, 4)
plt.plot(phi / math.pi, D)
```

```python
plt.polar(phi, A, 'r')
plt.polar(phi, B, 'b')
plt.polar(phi, C, 'g')
plt.polar(phi, D, 'y')
```
